### P1: Integración Númerica

Este problema consiste en resolver una integral de manera númerica.
Esta proviene de la ecuación de energía de un pendulo que se suelta del reposo en un ángulo $\phi_0$.

$$ 
E = −mgL cos \phi_0 = \frac{1}{2} mL^2 (\frac{d\phi}{dt})^2 − mgL cos \phi
$$

Así, se calcula el periodo despejando $dt$:

$$
dt = \sqrt(\frac{L}{g})\frac{d\phi}{\sqrt(2(cos\phi-cos\phi_0))}
$$

Esta última ecuación se puede integrar entre 0 y cuarto de periodo para llegar al ángulo $\phi_0$: 

$$
\frac{T}{4} = \sqrt(\frac{L}{g})\int^{\phi_0}_0\frac{d\phi}{\sqrt(2(cos\phi-cos\phi_0))}
$$

Finalmente, definienedo $T_0 = 2\pi \sqrt(L/g)$ que corresponde al periodo de un péndulo en el caso de pequeñaas amplitudes, se tiene:
$$
\frac{T}{T_0} = \sqrt(\frac{2}{\pi})\int^{\phi_0}_0\frac{d\phi}{\sqrt(2(cos\phi-cos\phi_0))}
$$

En primer lugar, definimos los parametros del problema:

```python
puntos = np.arange(0, 2* math.pi, 0.01)

phiCero = np.arange(0.01, math.pi/2, 0.01)


def f(phi):
    return 1/np.sqrt(2 * (math.cos(phi) - math.cos(phiCero)))
```

Notamos que la integral va a divergir ya que se tiene $cos\phi - cos\phi_0$. Por esto, se debe realizar un cambio de variable. Así, se utiliza el siguiente: 
$$
sen(\frac{\phi}{2}) = sen(\frac{\phi_0}{2})\cdot sen(\theta)
$$

Recordamos también que $cos\theta = 1 - 2sen^2\frac{\theta}{2}$

Así, se tiene:
$$
\begin{align}
cos\phi - cos\phi_0 = 1 - 2sen^2\frac{\phi}{2} - 1 + 2&sen^2\frac{\phi_0}{2} \\
= 2sen^2\frac{\phi_0}{2}- 2sen^2\frac{\phi}{2}
\end{align}
$$

Luego, por el cambio de variable:

$$
2sen^2\frac{\phi_0}{2}- 2sen^2\frac{\phi}{2} = 2sen^2\frac{\phi_0}{2} - 2sen^2(\frac{\phi_0}{2})\cdot sen^2(\theta)
$$

### P2: Resolución Númerica de una EDO

Este segundo problema busca implementar el método de Runge Kutta de orden 4 para la resolución del modelo de Lorenz.
Este modelo fue desarollado por Edward Lorenz con el objetivo de estudiar la convección atmosférica. Esto corresponde al resultado de la inestabilidad parcial del ambiente. Esto se traduce en fuertes corrientes de aire, tormentas eléctricas, tornados, etc... Así, Lorenz presentó un sistema de tres EDOs: 

$$
\frac{dx}{dt} = a(y-x)
$$


$$
\frac{dy}{dt} = x(b-z) - y
$$


$$
\frac{dz}{dt} = xy - cz
$$

Así, $x$ es proporcional al *ratio* de convección, $y$ es proporcional a la variación horizontal de la temperatura y $z$ lo es para la variación vertical de esta.

Sin embargo, el modelo se hizo famoso por presentar caos de una manera fácil de visualizar y entender. Esto quiere decir que para ciertos parametros ($a,b,c$) y condiciones iniciales el sistema presenta soluciones caóticas de la forma de *mariposa* (cf. https://en.wikipedia.org/wiki/Chaos_theory).

Los parametros a, b y c fueron introducidos como proporcionales a número de Prandtl, de Rayleigh y condiciones específicas de la capa atmosférica. Usualmente, se tiene que $a=10$, $c= 8/3$ y $b$ varía. Para nuestro estudio del modelo se tiene que $b=3R.RR$, donde $R.RR = 3.15$. Cabe destacar que se conoce que para $b=28$ el sistema presenta soluciones caóticas.

Para estudiar el problema, planteamos los parámetros de este:

```python
a = 10.0
c = 3/8
b = 33.15


def f(x, y):
    '''
    Funcion que define la primera ecuacion del sistema de Lorenz
    '''
    return a*(y - x)

def g(x, y, z):
    '''
    Funcion que define la segunda ecuacion del sistema de Lorenz
    '''
    return x*(b - z) - y

def e(x, y, z):
    '''
    Funcion que define la tercera ecuacion del sistema de Lorenz
    '''
    return x*y - c*z


valoresIniciales = [6.0 , 3.0, 6.0] # ultimos 6 a 4 numeros del rut

step = [0.01 , 0.1, 0.11, 0.12, 0.13, 0.14] # diferentes pasos de integracion

def puntosIntegracion(paso):
    puntos = []
    for i in paso:
        puntos.append(np.arange(0, 50, i))
  
    return puntos
```

Ahora, planteamos el método de Runge Kutta de orden 4:

Recordamos primero que de manera más simple el método se traduce al siguiente algoritmo: 

$${\dot {y}}=f(t,y),\quad y(t_{0})=y_{0}$$

Se tiene $y$ una función que depende del tiempo $t$ que nos interesa aproximar. Es conocido que ${\dot {y}}$ esta en funcion de t e y. Además, se conocen las condiciones iniciales de $y$, sea $y_0$, en $t_0$.

Con un paso de integración de h > 0  se define:
$$
{\displaystyle {\begin{aligned}y_{n+1}&=y_{n}+{\tfrac {1}{6}}\left(k_{1}+2k_{2}+2k_{3}+k_{4}\right),\\t_{n+1}&=t_{n}+h\\\end{aligned}}}
$$
para $n = 0, 1, 2, 3, ...$

$$
{\displaystyle {\begin{aligned}k_{1}&=h\ f(t_{n},y_{n}),\\k_{2}&=h\ f\left(t_{n}+{\frac {h}{2}},y_{n}+{\frac {k_{1}}{2}}\right),\\k_{3}&=h\ f\left(t_{n}+{\frac {h}{2}},y_{n}+{\frac {k_{2}}{2}}\right),\\k_{4}&=h\ f\left(t_{n}+h,y_{n}+k_{3}\right).\end{aligned}}}
$$

```python
def rungeKutta(f, g, e, iniciales, puntos):
    solve = []  
    # Multiple integration steps
    for x in puntos:
        # define initial conditions
        u1 = [iniciales[0]]
        u2 = [iniciales[1]]
        u3 = [iniciales[2]]
        
        # define step of integration
        h = x[1] - x[0]
        
        # RK4 on 3 variables
        for i in range(len(x) - 1): 
            kx1 = h*f(u1[i],u2[i])
            ky1 = h*g((u1[i]),u2[i],u3[i])
            kz1 = h*e(u1[i],u2[i],(u3[i]))
        
            kx2 = h*f(u1[i]+kx1/2, u2[i]+ky1/2)  
            ky2 = h*g(u1[i]+kx1/2, u2[i]+ky1/2, u3[i]+kz1/2) 
            kz2 = h*e(u1[i]+kx1/2, u2[i]+ky1/2, u3[i]+kz1/2)
    
            kx3 = h*f(u1[i]+kx2/2, u2[i]+ky2/2);
            ky3 = h*g(u1[i]+kx2/2, u2[i]+ky2/2, u3[i]+kz2/2) 
            kz3 = h*e(u1[i]+kx2/2, u2[i]+ky2/2, u3[i]+kz2/2)
            
            kx4 = h*f(u1[i]+kx3, u2[i]+ky3)
            ky4 = h*g(u1[i]+kx3, u2[i]+ky3, u3[i]+kz3)
            kz4 = h*e(u1[i]+kx3, u2[i]+ky3, u3[i]+kz3)
    
            u1.append(u1[i] + (kx1+(kx2*2)+(kx3*2)+kx4)/6)
            u2.append(u2[i] + (ky1+(ky2*2)+(ky3*2)+ky4)/6)
            u3.append(u3[i] + (kz1+(kz2*2)+(kz3*2)+kz4)/6)
        
        solve.append([u1, u2, u3])
    
    return solve
```

Luego, resolvemos para el sistema de Lorenz:

```python
res1 = rungeKutta(f, g, e, valoresIniciales, puntosIntegracion(step))
```

Finalmente, graficamos los resultados obtenidos. Primero, graficamos la resolución en 3 dimensiones. Luego, nos piden graficar $y$ versus $z$.

```python
def plotResRK(res= res1, step= step):
    
    for index, aRes in enumerate(res):
        fig = plt.figure()
        ax = fig.gca()
        ax.plot(aRes[1], aRes[2])
        ax.set_xlabel("Y")
        ax.set_ylabel("Z")
        ax.set_title("Resolución con un paso de " + str(step[index]))
        plt.show()

plotResRK()
```

Como podemos observar el cambio en el paso de integración hace de los valores que toma la función sean menos precisos.
Es decir, al tener un paso de integración mayor se *pierden* los valores de puntos que estén entre cada punto de integración. Esto se traduce a una función menos *suave* al momento de graficar. De manera más formal, la resolución del sistema se hace de manera menos inexacta.

Cabe destacar que para un paso de integración desde $ 0.14\leq h $ la integración diverge. Por esto el sistema nos advierte de un *valid value encountered in double_scalars* (Esto indica o una división por cero o *buffer overflow*). Claramente en este problema no dividimos por cero).

Finalmente, se estudia como varían las soluciones para condiciones iniciales similares:

```python
nuevasInit = []
for index, val in enumerate(valoresIniciales):
    if(index == 0):
        nuevasInit.append(val)
    nuevasInit.append(val)

res2 = rungeKutta(f, g, e, nuevasInit, puntos = puntosIntegracion(step))

def plotRes2(firstRes = res1, secRes= res2, puntos= puntosIntegracion(step)):
    fig = plt.figure()
    ax = fig.gca()
    ax.plot(puntos[0], firstRes[0][0])
    ax.set_xlabel("T")
    ax.set_ylabel("X")
    plt.show()
    
    fig = plt.figure()
    ax = fig.gca()
    ax.plot(puntos[0], secRes[0][0])
    ax.set_xlabel("T")
    ax.set_ylabel("X")
    plt.show()
    

plotRes2()
```
